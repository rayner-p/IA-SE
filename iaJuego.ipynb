```python
from tkinter import *
from tkinter import messagebox
from tkinter import ttk
import tkinter as tk
from neo4j import GraphDatabase
from neo4j import unit_of_work

uri = "neo4j://localhost:7687"
driver = GraphDatabase.driver(uri, auth=("neo4j", "neo4jj"))
#driver2 = GraphDatabase.driver("bolt://localhost:7687",  auth=("neo4j", "neo4jj"));
session = driver.session();
tx = session
def main():
    root= Tk()
    app = ventanaPrincipal(root)
class ventanaPrincipal:
    def __init__(self, master):
        self.master = master
        self.master.title('IA')
        self.master.geometry('00x500+0+0')
        self.master.config(bg = 'powder blue')
        self.frame =Frame(self.master, bg = 'powder blue')
        self.frame.pack()
        
        self.btnProducto = Button(self.frame, text='JUGAR', width= 17, command=self.ventanaNueva)
        self.btnProducto.grid(row=7, column = 0)
        
        self.btnCliente = Button(self.frame, text='RECOMENDACIONES', width= 17, command=self.ventanaNueva2)
        self.btnCliente.grid(row=7, column = 2) 
    def ventanaNueva(self):
        self.ventanaNueva = Toplevel(self.master)
        self.ventanaProducto = Producto(self.ventanaNueva) 
    def ventanaNueva2(self):
        self.ventanaNueva2 = Toplevel(self.master)
        self.ventanaCliente = Recomendaciones(self.ventanaNueva2) 
    def ventanaNueva3(self):
        self.ventanaNueva3 = Toplevel(self.master)
        self.ventanaPedido = Pedido(self.ventanaNueva3)
    
class Recomendaciones:  
  
    def __init__(self, window):
        self.wind = window
        self.wind.title('Recomendaciones') 
        frame = LabelFrame(self.wind)
        frame.grid(row = 0, column = 0, columnspan = 3, pady = 20)
        #ttk.Button(frame, text = 'RESTAURANTS', command=self.obtenerRestaurants).grid(row = 10, column = 2)  
        #ttk.Button(frame, text = 'HOTELES', command=self.realizarPedido).grid(row = 12, column= 3)  
        #ttk.Button(frame, text = 'SITIOS TURISTICOS', command=self.realizarPedido).grid(row = 12, column= 3)  

        btnR = Button(window, text = 'RESTAURANTS', command=self.obtenerRestaurants).grid(row = 10, column = 2)
        #btnH = Button(window, text = 'HOTELES').grid(row = 12, column= 3)       
        btnH = Button(window, text = 'SITIOS TURISTICOS').grid(row = 12, column= 3)      
    @unit_of_work(timeout=5)
    def obtenerRestaurants(self):
        friends = []
        
        result = tx.run("MATCH (n:Restaurant) RETURN n")
        for record in result:
            print(record)
            friends.append(record)
            return friends
    def crearMapa(self):
        query = "CALL gds.graph.create('k1Localidad',{Restaurant : {label: 'Restaurant',properties: 'gn'}},'*')"
    with driver.session() as session:
        friends = session.read_transaction(obtenerRestaurants)
        session.write_transaction(crearMapa)
        for friend in friends:
            print(friends)
    
    def obtenerCuentasCliente(self):
        messagebox.showwarning('Message title', 'Message content')  #shows warning message
                 
if __name__ == '__main__':
    main()
    window = Tk()
    window.geometry('400x200')
    #application = Cliente(window)
    window.mainloop()
```
